### Converting some strings to numeric variables

```python
# Remove symbols from Wage

FIFA_df["Wage"]=FIFA_df['Wage'].str.replace('€', '')
FIFA_df["Wage"]=FIFA_df['Wage'].str.replace('K', '')
FIFA_df["Wage"] = FIFA_df["Wage"].astype(int)
```

```python
# Remove symbols from Value

FIFA_df["Value"]=FIFA_df['Value'].str.replace('M', '')
FIFA_df["Value"]=FIFA_df['Value'].str.replace('€', '')
FIFA_df["Value"]=FIFA_df['Value'].str.replace('K', '')
FIFA_df["Value"] = FIFA_df["Value"].astype(float)
```

```python
# Remove symbols and transform height from feet to inches

FIFA_df["Height"]=FIFA_df["Height"].str.replace("'","")
FIFA_df['Height']  = FIFA_df['Height'].astype('string')

# Get the feet values
f = FIFA_df['Height'].str.slice(stop=1)
f = f.astype(int)

# Get the inches values
i = FIFA_df['Height'].str[1:]
i = i.astype(int)


FIFA_df['Height'] = f*12 + i
```

```python
# Remove symbols from Weight

FIFA_df["Weight"]=FIFA_df["Weight"].str.replace("lbs","")
FIFA_df["Weight"] = FIFA_df["Weight"].astype(int)
```

### One hot encoding

```python
dummies_position = pd.get_dummies(FIFA_df['Position'])
dummies_foot = pd.get_dummies(FIFA_df['Preferred Foot'])

FIFA_df = pd.concat([FIFA_df, dummies_position, dummies_foot], axis=1)
```

```python
# Drop original feature columns
FIFA_df = FIFA_df.drop(['Position', 'Preferred Foot'], axis=1)
```

```python
# Create train and test split

from sklearn.model_selection import train_test_split

#X   = FIFA_df[['Age', 'Overall', 'Potential', 'Value', 'International Reputation', 'Weak Foot', 'Skill Moves','Height', 'Weight', 'Crossing', 'Finishing', 'HeadingAccuracy', 'Volleys', 'Curve', 'FKAccuracy', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'Passing', 'Defender', 'Forward', 'GoalKeeper', 'MidFielder', 'Left', 'Right','Passing']] #get the input features
X   = FIFA_df[['Age', 'Overall', 'Potential', 'Value', 'International Reputation', 'Weak Foot', 'Skill Moves','Height', 'Weight', 'Crossing', 'Finishing', 'HeadingAccuracy', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes', 'Passing', 'Defender', 'Forward', 'GoalKeeper', 'MidFielder', 'Left', 'Right','Passing']] #get the input features

y   = FIFA_df['Wage']              #get the target

X_train, X_test, y_train, y_test = train_test_split(X,              #the input features
                                                    y,              #the label
                                                    test_size=0.3,  #set aside 30% of the data as the test set
                                                    random_state=7, #reproduce the results
                                                    stratify = FIFA_df[['Defender', 'Forward', 'GoalKeeper', 'MidFielder']])
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math

reg = LinearRegression().fit(X_train, y_train)
print('R squared training set for Linear Regression', round(reg.score(X_train, y_train),4))
print('R squared test set for Linear Regression', round(reg.score(X_test, y_test),4))
print('RMSE TRAIN', round(mean_squared_error(y_train, reg.predict(X_train), squared = False), 2))
print('RMSE TEST', round(mean_squared_error(y_test, reg.predict(X_test), squared = False), 2))
```

# Ridge Regression

```python
from numpy import arange
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
```

```python
# Define cross-validation method to evaluate model
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# Define model
model = RidgeCV(alphas=arange(0.01, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')

# Fit model
model.fit(X_train, y_train)

# Display lambda that produced the lowest test MSE
print(model.alpha_)
```

```python
# R-squared score
print('R squared training set for Ridge Regression', round(model.score(X_train, y_train), 4))
print('R squared test set for Ridge Regression', round(model.score(X_test, y_test), 4))
print('RMSE TRAIN', round(mean_squared_error(y_train, model.predict(X_train), squared = False), 2))
print('RMSE TEST', round(mean_squared_error(y_test, model.predict(X_test), squared = False), 2))
```

# LASSO

```python
from sklearn.linear_model import Lasso

# Define the model
lasso = Lasso(alpha=1.0)

# Fit the model
lasso.fit(X_train, y_train)

# R-squared score
print('R squared training set for LASSO', round(lasso.score(X_train, y_train), 4))
print('R squared test set for LASSO', round(lasso.score(X_test, y_test),4))
print('RMSE TRAIN', round(mean_squared_error(y_train, lasso.predict(X_train), squared = False), 2))
print('RMSE TEST', round(mean_squared_error(y_test, lasso.predict(X_test), squared = False), 2))
```

```python
from sklearn.linear_model import LassoCV

# Lasso Cross validation
lasso_cv = LassoCV(alphas = [0.0001, 0.001,0.01, 0.1, 1, 10], random_state=0).fit(X_train, y_train)

# R-squared score
print('R squared training set for CV LASSO', round(lasso_cv.score(X_train, y_train),4))
print('R squared test set for CV LASSO', round(lasso_cv.score(X_test, y_test),4))
print('RMSE TRAIN', round(mean_squared_error(y_train, lasso_cv.predict(X_train), squared = False), 2))
print('RMSE TEST', round(mean_squared_error(y_test, lasso_cv.predict(X_test), squared = False), 2))
```

In order to get the feature importance


```python
# Get the feature importance
feature_importance = pd.Series(index=X_train.columns, data=abs(lasso_cv.coef_))

# Sort the features by importance
sorted_feature_importance = feature_importance.sort_values(ascending=False)

# Plot the feature importance
plt.figure(figsize=(10, 5))
sorted_feature_importance.plot(kind='bar', color= 'green')
plt.title('Feature Importance')
plt.xlabel('Features')
plt.ylabel('Absolute Coefficient')
plt.show()
```
